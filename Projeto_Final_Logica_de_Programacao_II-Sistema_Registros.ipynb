# Sistema de Cadastro



```python
def main():
    global cadastros
    # AMBIENTE
    print('Boas vindas ao nosso sistema!')
    cadastros = ler_salvar_json(ler=True)

    # MENU
    opcao_sair = False
    while not opcao_sair:
        operacao_menu = print_menu()

        # OPERAÇÕES
        if operacao_menu == '1':
            entrada_new_user = add_usuario_input()
            processamento_de_user(usuario=entrada_new_user, cadastro_dict=cadastros)
        elif operacao_menu == '2':
            entrada = input_exclusão()
            excluir_usuarios(*entrada)
        elif operacao_menu == '3':
            edit_user()
        elif operacao_menu == '4':
            exibir_usuario( _user_info_input() )
        elif operacao_menu == '5':
            exibir_todos_usuarios()
        elif operacao_menu == '6':
            opcao_sair = True



    # FINALIZAÇÃO
    print('='*8, 'Finalizando programa...', '='*8)
    ler_salvar_json(ler=False, nomeArquivo='projetoModuloII.json', dados=cadastros)

#  ===========================================================================
# 0.1. LEITURA DE JSON
import json

def ler_salvar_json(ler=True, nomeArquivo='projetoModuloII.json', dados=None):
    if ler:
        with open(nomeArquivo) as file_json:
            cadastros = json.load(file_json)
            return cadastros
    else:
        # Criar o passo-a-passo para salvar o arquivo e não é necessário retornar nada
        with open(nomeArquivo, 'w') as file_json:
            json.dump(dados, file_json, ensure_ascii=False, indent=2)
            print('Arquivo JSON atualizado!')



# 0.2. MENU
def print_menu():
    menu_str = """\nMENU:

1 - Inserir usuário
2 - Excluir usuário
3 - Atualizar usuário
4 - Informações de um usuário
5 - Informações de todos os usuários
6 - Sair\n
"""
    valor_entrada = input(menu_str)

    # Tratamento de input inválido
    while valor_entrada not in list('123456'):
        valor_entrada = input('Valor de entrada errado. Insira uma opção válida:\n')
    return valor_entrada


# 1 - add
def add_usuario_input() -> dict:
    """Entrada do usuário para adicionar um novo usuário."""

    valor_entrada_nome = input("\nInsira o nome:\n")
    valor_entrada_telefone = input("\nInsira o telefone:\n")
    valor_entrada_endereco = input("\nInsira o endereco:\n")

    # Tratamento das entradas
    while valor_entrada_nome == '':
        valor_entrada_nome = input("\nNome inválido. Insira novamente o nome:\n")

    if valor_entrada_telefone == '':
        valor_entrada_telefone = 'Nao Informado'
    if valor_entrada_endereco == '':
        valor_entrada_endereco = 'Nao Informado'

    dict_input_user = {'Nome': valor_entrada_nome, 'Telefone': valor_entrada_telefone, 'Endereço': valor_entrada_endereco}
    return dict_input_user


def _registro_existente(usuario: dict, cadastro_dict: dict) -> bool:
    """No processamento é verificado se o usuário já existe no banco de dado,
    atribui um id sequencial e o status"""

    # Verifica a existência do cadastro
    for idx in cadastro_dict:
        if list(usuario.values()) == [cadastro_dict[idx]['Nome'], cadastro_dict[idx]['Telefone'], cadastro_dict[idx]['Endereço']]:
            # Alterar o status no cadastro
            cadastro_dict[idx]['Status'] = True
            print(f'Cadastro {idx} atualizado!')
            return True
    return False


def processamento_de_user(usuario: dict, cadastro_dict: dict):
    """No processamento é verificado se o usuário já existe no banco de dado,
    atribui um id sequencial e o status"""

    # Já existente
    if _registro_existente(usuario=usuario, cadastro_dict=cadastro_dict):
        return None

    # Criar um id sequencial
    novo_index = max([int(idx) for idx in cadastro_dict]) + 1

    # Modelagem do resultado - altera status caso já exista o nome
    usuario['Status'] = True

    cadastro_dict[str(novo_index)] = usuario
    print(f'\nNovo cadastro no sistema. (ID: {novo_index})')


# 2 - EXCLUSÃO
def input_exclusão():
    print('='*8, 'Exclusão de registro:', '='*8)
    entrada = input('\nInsira os IDs que deseja excluir (separados por espaço):\n').split(' ')
    return entrada

def excluir_usuarios(*ids):
    for id in ids:
        _excluir_logicamente_um_user(id)

def _excluir_logicamente_um_user(id):
    try:
        cadastros[str(id)]['Status'] = False
        print(f'ID {id} excluído!')
    except KeyError:
        print(f'ID {id} não existe na existe na base.\n')


# 3 - EDIÇÃO
def edit_user():
    print('='*8, 'Edição de registro:', '='*8)
    id = input('Insira o ID do usuário:\n')

    while id not in cadastros.keys():
        print('Usuário não encontrado!\n')
        id = input('Insira o ID do usuário:\n')

    alteracao_cod = input('\nQual informação deseja alterar?\n1 - Nome\n2 - Tefone\n3 - Endereço\n\n')

    if alteracao_cod == '1':
        novo_nome = input('Insira o nome:\n')
        cadastros[id]['Nome'] = novo_nome
    elif alteracao_cod == '2':
        novo_telefone = input('Insira o telefone:\n')
        cadastros[id]['Telefone'] = novo_telefone
    elif alteracao_cod == '3':
        novo_endereco = input('Insira o endereço:\n')
        cadastros[id]['Endereço'] = novo_endereco


# 4 - Informações de um usuário
def _user_info_input() -> list:
    # Lista de usuários para validação
    lista_usuarios = list(cadastros.keys())
    print('='*8, 'Informações de usuários', '='*8)

    # Verificação da entrada do usuário
    user_ids_valid = False
    while user_ids_valid == False:
        # user_ids_valid = _user_info_input(lista_usuarios)

        valor_entrada = input('\nInsira o ID do usuário: (para múltiplas consultas separe com espaço)\n')
        valor_entrada = valor_entrada.split(' ')

        # Validação da entrada
        if valor_entrada == []:
            continue

        # Validação da existência do id
        todos_existentes = True
        for i in valor_entrada:
            if i not in lista_usuarios:
                print(f'Usuário {i} não encontrado! Refaça a entrada.')
                todos_existentes = False
        if todos_existentes == False:
            continue
        elif todos_existentes == True:
            user_ids_valid = True

    return valor_entrada

def exibir_usuario(user_ids):
    for u in user_ids:
        user = cadastros[u]
        user_info_str = f"\nNome: {user['Nome']}\nTefone: {user['Telefone']}\nEndereço: {user['Endereço']}\n"
        print(user_info_str)
    return None


# 5 - Informações de todos usuários
# def exibir_todos_usuarios(cadastros: dict):
def exibir_todos_usuarios():
    lista_id_usuarios = list(cadastros.keys())
    print('='*8, 'Mostrando todos os registros ativos do cadastro:', '='*8)

    for user_id in lista_id_usuarios:
        user = cadastros[user_id]
        if user['Status'] == False:
            continue
        user_info_str = f"\nID: {user_id}\nNome: {user['Nome']}\nTefone: {user['Telefone']}\nEndereço: {user['Endereço']}\n"
        print(user_info_str)

    print('_'*40)
    return None
```

```python
# EXECUÇÃO DO PROGRAMA
main()
```
